```python
import pandas as pd
from IPython.display import display
from matplotlib.pylab import plt
```

```python
pf = pd.read_csv(r'D:\MySpider\SecondHandHouse\items.csv')
```

```python
pf.head()
```

```python
pf['total_price'] = pf['price']*pf['area']/10000
display(pf.head())
```

```python
pf_mean = pf.groupby('district')['price'].mean()
pf_mean.head(10)
```

```python
pf_count = pf.groupby('district')['price'].count()
pf_count.head(10)
```

```python
p2 = pf_mean.sort_values(ascending=False)
p2.head(10)
```

```python
plt.figure(figsize=(20,10))
plt.rc('font',family='SimHei',size=13)
plt.title(u'各区域平均房价')
plt.xlabel(u'北京区域')
plt.ylabel(u'平均房价')
plt.bar(pf_mean.index,pf_mean.values,color='g')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.rc('font',family='SimHei',size=13)
# plt.title(u'各区域二手房数量')
explode = [0]*len(pf_count)  # 每一块离开中心距离
explode[9] = 0.1
plt.pie(pf_count,radius=2,autopct='%1.f%%',shadow=True,labels=pf_count.index,explode=explode)
plt.axis('equal')
plt.show()
```
